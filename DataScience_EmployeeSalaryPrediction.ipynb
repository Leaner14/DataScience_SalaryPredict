# Encoding using label encoding

```python
from sklearn.preprocessing import LabelEncoder
import joblib

le_experience = LabelEncoder()
data['experience_level'] = le_experience.fit_transform(data['experience_level'])
joblib.dump(le_experience, "le_experience.pkl")

le_employment = LabelEncoder()
data['employment_type'] = le_employment.fit_transform(data['employment_type'])
joblib.dump(le_employment, "le_employment.pkl")

le_job_title = LabelEncoder()
data['job_title'] = le_job_title.fit_transform(data['job_title'])
joblib.dump(le_job_title, "le_job_title.pkl")



le_currency = LabelEncoder()
data['salary_currency'] = le_currency.fit_transform(data['salary_currency'])
joblib.dump(le_currency, "le_currency.pkl")

le_residence = LabelEncoder()
data['employee_residence'] = le_residence.fit_transform(data['employee_residence'])
joblib.dump(le_residence, "le_residence.pkl")

le_location = LabelEncoder()
data['company_location'] = le_location.fit_transform(data['company_location'])
joblib.dump(le_location, "le_location.pkl")

le_size = LabelEncoder()
data['company_size'] = le_size.fit_transform(data['company_size'])
joblib.dump(le_size, "le_company_size.pkl")


data

```

```python
print(data['job_title'].unique()[:5])
```

```python
X = data.drop('salary_in_usd', axis=1)
y = data['salary_in_usd']
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
xtrain
```

#Model used:GradientBoostingRegresso

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor()
model.fit(xtrain, ytrain)
predict = model.predict(xtest)
predict
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(ytest, ypred)
rmse = np.sqrt(mean_squared_error(ytest, ypred))
r2 = r2_score(ytest, ypred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R² Score:", r2)
```

<h3>Graph: to check variation for Actual vs predicted salary</h3>

```python
import matplotlib.pyplot as plt

plt.scatter(ytest, ypred)
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Actual vs Predicted Salary")
plt.plot([ytest.min(), ytest.max()], [ytest.min(), ytest.max()], 'r--')  # perfect fit line
plt.show()
```

#cross-validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, xtrain, ytrain, cv=5, scoring='r2')
print("Cross-validated R²:", scores.mean())
```

<h3> =>calculates the residuals, the differences between the actual values (ytest) and predicted values (ypred).</h3>



```python
import matplotlib.pyplot as plt
residuals = ytest - ypred
plt.hist(residuals, bins=30)
plt.title("Distribution of Prediction Errors (Residuals)")
plt.xlabel("Error")
plt.ylabel("Frequency")
plt.show()
```

# UI: Using streamlite

```python
%%writefile app.py
import streamlit as st
import numpy as np
import joblib

# Load model
model = joblib.load("salary_model.pkl")

# Load encoders
le_experience = joblib.load("le_experience.pkl")
le_employment = joblib.load("le_employment.pkl")
le_job_title = joblib.load("le_job_title.pkl")

# le_salary_title = joblib.load("le_salary_title.pkl")

le_currency = joblib.load("le_currency.pkl")
le_residence = joblib.load("le_residence.pkl")
le_location = joblib.load("le_location.pkl")
le_size = joblib.load("le_company_size.pkl")

# Streamlit UI
st.title("💼 Data Science Salary Predictor")

work_year = st.selectbox("Work Year", [2020, 2021, 2022, 2023, 2024, 2025])
experience_level = st.selectbox("Experience Level", le_experience.classes_)
employment_type = st.selectbox("Employment Type", le_employment.classes_)
job_title = st.selectbox("Job Title", le_job_title.classes_)

salary = st.number_input("Salary",min_value=0.0, step=1000.0)

salary_currency = st.selectbox("Salary Currency", le_currency.classes_)
employee_residence = st.selectbox("Employee Residence", le_residence.classes_)
remote_ratio = st.slider("Remote Work Ratio (%)", 0, 100, 50)
company_location = st.selectbox("Company Location", le_location.classes_)
company_size = st.selectbox("Company Size", le_size.classes_)

# Encode inputs
encoded_experience = le_experience.transform([experience_level])[0]
encoded_employment = le_employment.transform([employment_type])[0]
encoded_job_title = le_job_title.transform([job_title])[0]


encoded_currency = le_currency.transform([salary_currency])[0]
encoded_residence = le_residence.transform([employee_residence])[0]
encoded_location = le_location.transform([company_location])[0]
encoded_size = le_size.transform([company_size])[0]

input_data = np.array([[work_year, encoded_experience, encoded_employment,
                        encoded_job_title,salary, encoded_currency, encoded_residence,
                        remote_ratio, encoded_location, encoded_size]])

# Predict
if st.button("Predict Salary"):
    prediction = model.predict(input_data)[0]
    st.success(f"💰 Predicted Salary (USD): ${prediction:,.2f}")
```
